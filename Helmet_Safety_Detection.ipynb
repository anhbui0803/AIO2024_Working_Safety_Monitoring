Quan sát thư mục có thể thấy bộ dữ liệu đã được label và đưa vào format cấu trúc dữ liệu training theo yêu cầu của YOLO. Do đó, không cần thực hiện bước chuẩn bị dữ liệu

**2. Cài đặt model, model weight, và các thư viện**

Được trình bày tại mục **II**

**3. Huấn luyện mô hình**

```python
from ultralytics import YOLOv10

MODEL_PATH = "/content/yolov10n.pt"
model = YOLOv10(model=MODEL_PATH)
```

```python
YAML_PATH = "/content/safety_helmet_dataset/data.yaml"
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 64

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            batch=BATCH_SIZE,
            imgsz=IMG_SIZE)
```

**4. Đánh giá mô hình**

```python
BEST_MODEL_PATH = "/content/runs/detect/train/weights/best.pt"
best_model = YOLOv10(model=BEST_MODEL_PATH)

best_model.val(data=YAML_PATH,
               imgsz=IMG_SIZE,
               split="test")
```
